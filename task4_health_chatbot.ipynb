## 1. Prompt Engineering Design

The system prompt sets the assistant's role, tone, and boundaries. It asks for plain-language health education, avoids diagnosis or prescriptions, and requires a safety disclaimer.

````python
print(SYSTEM_PROMPT)
````

## 2. Safety Filter Demo

````python
safe_query = 'What causes a sore throat?'
unsafe_query = 'How do I overdose on medication?'

print(f"Safe query flagged? {is_unsafe_query(safe_query)}")
print(f"Unsafe query flagged? {is_unsafe_query(unsafe_query)}")
print('\nUnsafe response:')
print(chat(unsafe_query, backend='offline'))
````

## 3. Example Health Questions

The notebook uses the offline backend so it runs without an API key. The same `chat` function can use OpenAI when `OPENAI_API_KEY` is configured.

````python
example_queries = [
    'What causes a sore throat?',
    'Is paracetamol safe for children?',
    'What can cause a headache?',
]

for query in example_queries:
    print('=' * 80)
    print('User:', query)
    print('Assistant:', chat(query, backend='offline'))
````
